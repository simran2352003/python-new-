```python
text = " this is a simple NLP pipeline using NLTK liberary"
```

```python
tokens = word_tokenize(text)
```

```python
tokens = [w for w in tokens if w. lower() not in stopwords.words('english')]
```

```python
ps = PorterStemmer()
```

```python
tokens = [ps.stem(w) for w in tokens]
```

```python
vector = [len(w) for w in tokens]
```

```python
max_len = 10
```

```python
padded = vector + [0]*(max_len - len(vector))
```

```python
print("final output:", padded[:max_len])
```
